### Printing Strings

Strings are line(s) of text that the computer interpets as words. The syntax for strings in Python is using either **double quotes** (" ") or **single quotes** (' '). When the computer reads in the first quotation mark, it will interpet everything after that as a string, until it hits a second quotation mark. Let's write our very first program.

One of the most basic programs to tell the computer to print 'Hello World'. To do this, we'll make use of Python's internal function **print**. 

```python
print('Hello world')
```

And there you have it! You have coded up your very first program! See that the color of the text within the quotation marks turned a different color, indicating that it will be interpeted as a string

Try printing out another sentence

Now lets look at some potential bugs of strings.
Try running the following cell.

```python
print('Hello world, it's a nice day out')
```

Uh-oh, what happened with the example above? The computer read that first (single) quotation mark and looked for the next quotation mark (which in our eyes is an apostrophe) as the end of the string. Then it reads the words after as variables (note the color change) and hits the last quotation mark where it then looks for an ending quotation mark (but sadly never finds it). When the computer reaches a quotation mark, either single or double, it will end the string at the next iteration of that mark. To avoid this we could write it as

```python
print("Hello world, it's a nice day out")
```

There we are. In the above case, the computer read the double quotation mark and looked for the ending double quotation mark - ignoring the apostrophe. Use the next few black cells to play around with the **print**  function.

## Creating Functions

Scientists and engineers deal with equations all the time. While we can use the math notation we learned above to code up functions, what if we need to use a particular function five, ten, a hundred times? It would get old really fast to code that particular function x amount of times. Just like variables, we can define functions at the beinning of our code and use them throughout. Let's code the function for a free falling object:

$y = -\frac{1}{2}gt^2+y_0$

The first thing one needs to figure out is what are our function's input? In the case of a free falling object, the two input variables are time **t** and initial height __y0__

```python
def freeFall(t,y0=0):
    '''
    Calculates how far an object dropped after time t has passed
    
    t -> time (int or list)
    y0 -> initial height (int)
    
    '''
    
    g = 9.81 # dgravitional constant on Earth
    
    y = -0.5*g*t**2+y0
    
    print(y)
    
    return None
```

Lets breakdown what you see above. First, one starts every function with the syntax **def**. This tells the computer that you are defining a function. It is then followed by the function name which, like variable names, should be short but practical. After the function name are parenthesis which hold the input variables. 

Notice that y0 has a default value, but t doesn't. This lets us have y0 in our function, but default it to being dropped from ground level.

Following the input variables, there is a semicolon which is important. The coding language C (which is what Python is built from) uses {} to "wrap" function code:

int freeFall(parameters){

    insert code
    
    return stuff
    
}

Anything wrapped inside the {} will be apart of the function. Python is similar but drops the {} and begins the function with a semicolon. The other key feature is that all the "body code" (code within the function) is indented beneath the function name. This is how Python keeps track of function code vs regular code. 

The first thing I put in the function is a small description of what it does, what each variable means, and what type of input they can be. This is useful down the line if you forget what a certain function does or what data type its inputs are suppose to be, you can just check it by:

```python
? freeFall
```

You can do that for any function/variable. Something important to note is that function descriptions can only be defined with multi-line comments (''' ''') and not single-line comments (#).


After the short description, I defined g followed by a comment of what g is. In this case, g is a **local variable**. We could define g outside of the function which would make it a **global variable**, but this is particularly dangerous. Let's look at an example where I define g, then define my function and use it, then *redefine* g and use the function again:

```python
g = 9.81

def testFunction(t,y0=0):
    
    y = -0.5*g*t**2+y0
    
    return y


print(testFunction(10))

g = 20

print(testFunction(10))
```

As you see, this ends up altering our given function. To avoid this type of error, make constants used in a function local variables. If you're **ABSOLUTELY** sure that a constant/variable name will not be altered then feel free to make it a global variable, but just be careful.

### More detail on global and local variables

Variables defined inside functions are known as local variables (i.e., they are local to only the function). Local variables cannot be used outside of a function unless you return that variable. 

```python
def randomFunction():
    '''
    Used to illustrate local variables
    '''
    
    local_variable = "this is a local variable"
    print(local_variable)
    
    return None

randomFunction()
print(local_variable)
```

I created a function that prints out the local_variable defined inside it but when I try to print local_variable, it fails because there is no variable named local_variable saved to memory. What if we define a local variable the same name as a global variable?

```python
foo = 'Global Foo'

def randomFunction():
    
    foo = 'Local Foo'
    print(foo)
    
    return None

print(foo)
randomFunction()
```

The function only recognizes the local variable foo. If for some reason you wanted to use the global foo inside the function, you just need to specify that

```python
foo = 'Global Foo'

def randomFunction():
    global foo
    print(foo)
    foo = 'Local Foo'
    print(foo)
    
    return None

randomFunction()
```

## For Loops

Let's combine everything we've learned so far into one of the fundamental tools in Python - the __For Loop__. Again, returning to the 1D projectile problem, say instead of wanting to find the position of an object went at one instance of time, we wanted to get a whole _list_ of positions at various times. We could code each individual instance line by line and run it, but that would be tedious and inefficient. This is where the __For Loop__ comes in, it will allow us to perform the same calculation while iterating through values. 

```python
positions = [] # This creates an empty list to hold our position values 
time = [] # This will hold time values

g = 9.81 # gravitational constant on Earth

'''
The general structure of the loop is:

for i *insert condition*:
    *loop code*
    
- i is just a dummy variable used to keep track of the current value being used. i,j, and k are usual dummy variables used. 
- The condition comes in many different flavors. For now, we'll work with the simplist case - iterating up to a number
- The loop code is exactly that, the code/algorithm you want to perform. This code will be tabbed under the *for* command. 

In our case, it will be an object dropping from rest. Let us look at the first 20 seconds.

'''
for i in range(21): 
    y = -0.5*g*i**2 # calculates the position
    time.append(i) # adds time value to list
    positions.append(y) # adds position value to the end of the list
    
print(positions)
```

Let's go more indepth of the for loop structure.

- We start a for loop with the __for__ command followed by a dummy variable, in this case i (this dummy variable can be anything, but it's just convention to use i,j,k as in math). 
- __in range(21)__ tells the computer that we want values of i within range of 21, but not 21 itself. Remember that Python starts its indexing at 0 so if we want the first 20 seconds, we need to go to 21
- We end our loop with :. This tells the computer that our conditions are done
- The 'loop code' is then tabbed under the loop. Any code that is tabbed under the __for loop__ will be considered inside the loop
- The loop ends once our conditions are met, in this case when i = 21

This is a very basic case and __for loops__ can become VERY intricate and complex depending on the calculation. The conditional statement can be more complex, you can have *nested for loops* which are for loops within a for loop (which is a rabbit hole that one can fall in - trust me on this one), your dummy variable can actually iterate through a list of values, etc. The possibilities are endless. This is why comments almost become necessary for one to follow the code. 

## Libraries and packages

Well great we have our list of positions, but that's no help to us if we can't visualize any of our data. Say we want to plot position versus time, if we were in a low level coding language such as C, we'd have to code it ourselves. This is where Python (a high level code) really shines, we can call on a package of code from a library where the functions are already coded up and we can make immediate use of it. 

__Libraries__ are exactly what they are named - it is a collection of code already made. __Packages__ are the "books" of the library - usually a lump of code that all fall together in one way or another. __Functions__ within the package can be thought up as chapters within a book. 

Making use of __libraries, packages, and functions__ are analogous to going to a real library and checking out a book. If one is writing a research paper, they go to the library and check out one or two books. And usually only a few selective chapters within those books are relevant to the paper.

Similarly, when writing code in Python one "calls" the specific library and package needed (sometimes even specifying the specific function). To "call" a library, one uses the __import__ command, then they specify which library. One specifies the package by adding a period after the library then the package name (ie __matplotlib.pyplot__). I'll provide a (far from exhaustive) list of common libraries used:

- matplotlib.pyplot -> used for all of ones plotting needs
- numpy -> a standard for doing maths as their functions are usualy MUCH faster than Python's internal
- scipy -> contains scientific functions/algorithms
- pandas -> a library that is very helpful in handling and organizing large amount of data
